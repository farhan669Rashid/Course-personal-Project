```python
import numpy as np
import pandas as pd

DATE_COL  = "filing_date"
VALUE_COL = "value"

assert DATE_COL in df.columns,  f"Column {DATE_COL} not found in df"
assert VALUE_COL in df.columns, f"Column {VALUE_COL} not found in df"

df_units = df.copy()
df_units[DATE_COL] = pd.to_datetime(df_units[DATE_COL])

# 1) Aggregate total value by filing_date (may contain mixed units: thousands vs dollars)
g = df_units.groupby(DATE_COL)[VALUE_COL].sum().sort_index()

# 2) Detect scale difference by comparing log10 values (bimodal distribution detection)
logv = np.log10(g.values + 1e-9)
median = np.median(logv)
low_mask = logv < median
high_mask = ~low_mask

gap = (logv[high_mask].mean() - logv[low_mask].mean()) if (low_mask.any() and high_mask.any()) else 0.0

print(f"[Unit Detection] Mean log10 gap between groups: {gap:.2f} (≥2 indicates ~100x scale difference)")

# 3) Identify dates reported in thousands of USD
thousands_dates = []
if gap > 2.0:
    thousands_dates = g.index[low_mask]
else:
    # If no clear bimodality, fall back to detecting any >100x jump between consecutive periods
    ratio = g.iloc[1:].values / (g.iloc[:-1].values + 1e-12)
    jump_idx = np.where(ratio > 100)[0]
    thousands_dates = g.index[: jump_idx[0]+1] if len(jump_idx) else []

print("Number of filing_dates identified as 'thousands USD' unit:", len(thousands_dates))
if len(thousands_dates):
    print("Range:", thousands_dates.min(), "→", thousands_dates.max())

# 4) Multiply 'value' by 1000 for those periods to unify all into USD
df_units.loc[df_units[DATE_COL].isin(thousands_dates), VALUE_COL] = (
    df_units.loc[df_units[DATE_COL].isin(thousands_dates), VALUE_COL] * 1000.0
)

# 5) Sanity check: after correction, no ~1000x jumps should remain between consecutive periods
g_fixed = df_units.groupby(DATE_COL)[VALUE_COL].sum().sort_index()
ratio_fixed = g_fixed.iloc[1:].values / (g_fixed.iloc[:-1].values + 1e-12)
print("Max consecutive-period jump ratio after correction:", float(ratio_fixed.max()))

# 6) Replace original df with corrected version for subsequent sector aggregation and trajectory building
df = df_units
```

# Build traj

### The AR(1) model serves a critical psychological purpose in this Inverse Reinforcement Learning (IRL) framework: It represents the Fund Manager's "Expectation" or "Belief."

```python
# ---------- deps ----------
import numpy as np
import pandas as pd
import torch
import yfinance as yf
import re

# ---------- constants ----------
GICS_ORDER = [                  # Fixed order of the 11 GICS sectors (as column order/dimension)
    "Energy","Materials","Industrials","Consumer Discretionary","Consumer Staples",
    "Health Care","Financials","Information Technology","Communication Services",
    "Utilities","Real Estate"
]
SECTOR_TICKERS = {              # Common ETF codes corresponding to 11 industries (used to approximate industry returns)
    "Energy":"XLE","Materials":"XLB","Industrials":"XLI","Consumer Discretionary":"XLY",
    "Consumer Staples":"XLP","Health Care":"XLV","Financials":"XLF","Information Technology":"XLK",
    "Communication Services":"XLC","Utilities":"XLU","Real Estate":"XLRE",
}

# ---------- helpers ----------

# Normalize the input (number/string) into a legal industry name
def _to_gics_name(s):
    if pd.isna(s): return None  # Missing values return None directly (indicating that they cannot be mapped)
    if isinstance(s,(int,np.integer)) and 0<=int(s)<11:
        return GICS_ORDER[int(s)]  # If it is a number between 0 and 10, the subscript is the industry name
    s = re.sub(r'\s+',' ',str(s)).strip().lower()  # Unify uppercase and lowercase letters and spaces
    for name in GICS_ORDER:
        if s == name.lower():   # Returns an exact match to the standard industry name (lowercase)
            return name
    return None

# Find the latest date <= target in time index idx
def _nearest(idx, target):
    pos = idx.searchsorted(pd.to_datetime(target), side="right") - 1  # Binary search positioning
    pos = max(0, min(pos, len(idx)-1))  # Boundary protection, guaranteed to be in [0, len-1]
    return idx[pos]             # Return the most recent valid date

# Compatible with yfinance
def _pick_close(df):
    if isinstance(df.columns, pd.MultiIndex):
        lvl0 = df.columns.get_level_values(0)
        key = "Adj Close" if "Adj Close" in lvl0 else "Close"  # Priority adjustment closing
        return df[key]
    else:
        cols = df.columns.tolist()
        key = "Adj Close" if "Adj Close" in cols else "Close"
        return df[key]

# Download price series for industry ETFs and SPY
def _fetch_sector_spy_prices(periods, pad_days=7):
    start = pd.to_datetime(min(periods)) - pd.Timedelta(days=pad_days)  # Start a little early
    end   = pd.to_datetime(max(periods)) + pd.Timedelta(days=pad_days)  # Delay the end

    raw = yf.download(" ".join(SECTOR_TICKERS.values()),
                      start=start, end=end, auto_adjust=False, progress=False)  # Pull industry ETFs
    px = _pick_close(raw)                     # Get the adjusted/closing price subtable
    if isinstance(px, pd.Series): px = px.to_frame()  # Convert a single column into a two-dimensional table for easy subsequent processing
    px = (px.rename(columns={v:k for k,v in SECTOR_TICKERS.items()})     # Column name changed from ETF to industry name
            .reindex(columns=GICS_ORDER)                                 # Sort columns by fixed industry order
            .sort_index().ffill().bfill())                               # Sort by date and fill in missing values

    spy_raw = yf.download("SPY", start=start, end=end,
                          auto_adjust=False, progress=False)              # Download SPY prices
    spy = _pick_close(spy_raw).sort_index().ffill().bfill()              # Same
    return px, spy                   # Industry Price List & SPY Price Series

def _sector_simple_returns_between_filings(periods, sector_px, spy_px=None, excess=True):
    rows = []                       # Save the 11-dimensional simple benefits of each interval
    for t in range(len(periods)-1): # Traverse adjacent disclosure periods (t -> t+1)
        t0, t1 = _nearest(sector_px.index, periods[t]), _nearest(sector_px.index, periods[t+1])  # Align to trading day
        s0, s1 = sector_px.loc[t0], sector_px.loc[t1]  # Take the industry prices at both ends of the interval
        r = (s1/s0 - 1.0).astype(float).replace([np.inf,-np.inf], np.nan).fillna(0.0)  # Calculate simple benefits and clean up
        if excess and spy_px is not None:            # If you want excess returns (relative to SPY)
            b0 = float(spy_px.loc[_nearest(spy_px.index, periods[t])])      # Range starting point SPY
            b1 = float(spy_px.loc[_nearest(spy_px.index, periods[t+1])])    # End point of the interval SPY
            r = r - (b1/b0 - 1.0)                 # Industry Return - SPY Return = Excess
        rows.append(r.values.astype(np.float32))
    return np.stack(rows, axis=0)  # [T-1,11]

def _estimate_ar1(R_obs, ridge=1e-6):  # Estimate industry-by-industry AR(1) parameters using observed returns
    """Industry by industry AR(1): r_t = phi * r_{t-1} + eps"""
    if R_obs.shape[0] < 3:              # Give a gentle default value when there are too few samples
        phi = np.zeros(11, dtype=np.float32)
        Sigma = 1e-5 * np.eye(11, dtype=np.float32)  # The residual covariance takes a very small diagonal matrix
        return phi, Sigma
    Y, X = R_obs[1:], R_obs[:-1]        # Y is the current return, X is the return lagged by one period
    phi = np.zeros(11, dtype=np.float32); resid = np.zeros_like(Y)  # Pre-allocation coefficients and residuals
    for i in range(11):                 # Separate regression for each industry
        num = float(np.dot(Y[:,i], X[:,i])); den = float(np.dot(X[:,i], X[:,i]) + ridge)  # Least squares (with ridge term)
        phi[i] = num/den                # Slope coefficient phi_i
        resid[:,i] = Y[:,i] - phi[i]*X[:,i]   # Calculate the residual sequence
    Sigma = np.cov(resid, rowvar=False)       # Covariance matrix of residuals
    Sigma = 0.5*Sigma + 0.5*np.diag(np.diag(Sigma)) + 1e-8*np.eye(11)  # Stabilization (shrink to the diagonal)
    return phi, Sigma                  # Return AR(1) coefficients and residual covariances

def _simulate_ar1(Tm1, phi, Sigma, r0=None, seed=None, clip=(-0.99, 3.0)):  # Sample a return path according to AR(1)
    rng = np.random.default_rng(seed)  # Random number engine (seed reproducible)
    R = np.zeros((Tm1,11), dtype=np.float32)  # Preallocation [T-1,11]
    R[0] = np.zeros(11, dtype=np.float32) if r0 is None else np.asarray(r0, dtype=np.float32)  # Initial r_0
    for t in range(1, Tm1):
        eps = rng.multivariate_normal(np.zeros(11), Sigma).astype(np.float32)  # Sampling residual ~ N(0, Sigma)
        R[t] = phi * R[t-1] + eps       # r_t = phi * r_{t-1} + eps
    if clip is not None:
        lo, hi = clip
        R = np.clip(R, lo, hi)          # Truncate extreme values to ensure 1+r > 0
    return torch.tensor(R, dtype=torch.float32)  # Convert to tensor

# ---------- main: df -> traj (AR1) ----------
def build_sector_traj_from_df_AR1(df_raw, sector_map, value_unit=1000.0, device="cpu",
                                  min_shares=1e-6, use_excess=True, seed=42):
    """
    Return traj：
      x_t      [T,11] Industry USD holdings
      r_t      [T-1,11] Realized simple industry returns (here "real raw returns", used for calculating u_t)
      bar_r_t  [T-1,11] Expected return bar_r_t = Phi * r_{t-1} (estimated based on the sequence specified by use_excess)
      Sigma_r  [11,11]  Residual covariance estimation
      u_t      [T-1,11] Net USD trading volume (calculate according to MDP PDF)
      C_t      [T-1,1]  net cash flow
      B_t      [T-1,1]  Benchmark
      cash_t   [T,1]    Cash Processing (USD)
    """
    # FORCE DOUBLE PRECISION (Float64) for Financial Math
    dtype = torch.float64

    df = df_raw.copy()
    # check columns
    need = {'cusip','value','shares','filing_date'}
    if not need.issubset(df.columns):
        raise ValueError(f"df_raw need columns: {need}")
    df['value']  = pd.to_numeric(df['value'],  errors='coerce').fillna(0.0)          # convert to number
    df['shares'] = pd.to_numeric(df['shares'], errors='coerce').fillna(0.0)          # convert to number
    df['filing_date'] = pd.to_datetime(df['filing_date'])                            # convert to datetime

    # Sector Map to 11 Industries
    m = dict(sector_map[['cusip','sector']].values) if isinstance(sector_map,pd.DataFrame) else dict(sector_map)
    df['sector'] = df['cusip'].map(m).map(_to_gics_name)                              # Get standardized industry name based on mapping
    df = df.dropna(subset=['sector']).copy()                                          # Discard rows that cannot be mapped to an industry(can be adjusted)

    periods = sorted(df['filing_date'].unique().tolist())                             # Take all disclosure periods and sort them
    if len(periods) < 2:                                                              # Need at least two periods
        raise ValueError("At least two issues of 13F are required to construct the trajectory.")

    # x_t
    val_sector = (df.groupby(['filing_date','sector'])['value']                       # Aggregate value by (period, industry) (unit: thousands of US dollars)
                    .sum().unstack('sector')                                          # Transferred industries
                    .reindex(index=periods, columns=GICS_ORDER, fill_value=0.0)       # Align rows and columns to a fixed order and fill with zeros
                    .sort_index())                                                    # Sort by date
    X_dollar = (val_sector * float(value_unit)).astype(float).replace([np.inf,-np.inf], np.nan).fillna(0.0)  # Multiply units to convert to dollars and clean up exceptions
    X = torch.tensor(X_dollar.values, dtype=dtype, device=device)             # [T,11] Industry USD holdings tensor
    AUM = X_dollar.sum(axis=1).astype(float)                                          # [T] Total assets per period (USD)

    # ------- Separate "real raw returns for transfers" and "series for estimates/expectations" -------
    sector_px, spy_px = _fetch_sector_spy_prices(periods)                             # Download sector ETF and SPY prices

    # (1) r_t_real_raw：True "raw" industry returns (not excess), driving transfers and resolving u_t
    R_real_np = _sector_simple_returns_between_filings(                               # Calculate the raw industry returns for adjacent disclosure periods
        periods, sector_px[GICS_ORDER], spy_px, excess=False                          # excess=False means no SPY reduction
    )
    # assure it is float64
    R_real = torch.tensor(R_real_np, dtype=dtype, device=device)              # Convert to tensor [T-1,11] as realized r_t

    # (2) R_for_model：Series to use for estimating AR(1) (optional: excess or original, controlled by use_excess)
    R_for_model_np = _sector_simple_returns_between_filings(                           # The second set of return series is only used for modeling
        periods, sector_px[GICS_ORDER], spy_px, excess=use_excess                     # True=Excess, False=Original
    )
    phi, Sigma = _estimate_ar1(R_for_model_np)                                        # Use this sequence to estimate the AR(1) coefficient and residual covariance
    # assure it is float64
    phi_t = torch.tensor(phi, dtype=dtype, device=device)                     # Coefficients converted to tensors (11,)

    # bar_r_t = Phi * r_{t-1}
    bar_R = torch.zeros_like(R_real, dtype=dtype)
    bar_R[0] = 0.0                                                                    # The first has no previous value, so set to 0
    bar_R[1:] = torch.tensor(R_for_model_np[:-1], dtype=dtype, device=device) * phi_t  # Sector by sector Φ r_{t-1}
    # assure it is float64
    Sigma_t = torch.tensor(Sigma, dtype=dtype, device=device)                 # Covariance Matrix Tensorization [11,11]

    # ------- Use "real original income" to reverse solve u_t -------
    den = (1.0 + R_real).clamp_min(1e-6)
    U = (X[1:] / den) - X[:-1]
    C = U.sum(dim=1, keepdim=True)
    B = torch.zeros((len(periods)-1, 1), dtype=dtype, device=device)          # Benchmark[T-1,1]

    # Cash
    cash_usd = [torch.tensor([0.0], dtype=dtype, device=device)]              # Initial cash (USD), set to 0 here
    cost_rate = 1e-3                                                                   # Transaction fee rate (e.g. 0.1% = 10bps)
    for t in range(len(periods)-1):                                                    # Update cash balance period by period
        traded_dollars = torch.linalg.vector_norm(U[t], ord=1)                         # Approximate transaction amount: sum of L1 norm
        trade_cost_usd = cost_rate * traded_dollars                                    # Transaction costs (USD)
        cash_next_usd = torch.clamp(cash_usd[-1] + C[t] - trade_cost_usd, min=0.0)     # Cash = Cash + Net Inflow - Cost, and is not negative
        cash_usd.append(cash_next_usd)
    CASH_USD = torch.stack(cash_usd, dim=0)                                            # [T,1]

    # Clean up(reset NaN/Inf to zero to ensure stability of subsequent training)
    X        = torch.nan_to_num(X,        nan=0.0, posinf=0.0, neginf=0.0)
    R_real   = torch.nan_to_num(R_real,   nan=0.0, posinf=0.0, neginf=0.0)
    bar_R    = torch.nan_to_num(bar_R,    nan=0.0, posinf=0.0, neginf=0.0)
    U        = torch.nan_to_num(U,        nan=0.0, posinf=0.0, neginf=0.0)
    C        = torch.nan_to_num(C,        nan=0.0, posinf=0.0, neginf=0.0)
    CASH_USD = torch.nan_to_num(CASH_USD, nan=0.0, posinf=0.0, neginf=0.0)

    return {
        "x_t": X,                                                                      # [T,11]
        "r_t": R_real,                                                                 # [T-1,11]
        "bar_r_t": bar_R,                                                              #（Φ r_{t-1})[T-1,11]
        "Sigma_r": Sigma_t,                                                            # [11,11]
        "u_t": U, "B_t": B, "C_t": C, "cash_t": CASH_USD,                               # action/benchmark/net_cash_flow/cash state(start from 0)
        "periods": periods, "sectors": GICS_ORDER,                                     # time point and industry order
        "ar1": {"phi": torch.tensor(phi)}                                              # AR(1) coefficient
    }


# ---------- small utility: traj -> states ----------
def build_states_from_traj(traj):  # Split traj into a list of "step-by-step states" (for easier feeding to the model)
    """
    Returns a list of length T-1, each element is a state s_t:
      {'x': [11], 'bar_r': [11], 'Sigma_r': [11,11], 'C': [1]}
    """
    X, barR, C, Sigma = traj["x_t"], traj["bar_r_t"], traj["C_t"], traj["Sigma_r"]  # Extract the required fields
    T = X.shape[0]                # Total number of periods T
    states = []                   # The list of states that will be returned
    for t in range(T-1):          # Only until T-1 (each state is used in the interval t->t+1)
        states.append({"x": X[t], "bar_r": barR[t], "Sigma_r": Sigma, "C": C[t]})  # Pack the features of each step into a dictionary
    return states                 # Return a chronological sequence of states
```

```python
# ---------- main: df -> traj (AR1) ----------
def build_sector_traj_from_df_AR1_old(df_raw, sector_map, value_unit=1000.0, device="cpu",
                                  min_shares=1e-6, use_excess=True, seed=42):
    """
    Return traj：
      x_t      [T,11] Industry USD holdings
      r_t      [T-1,11] Realized simple industry returns (here "real raw returns", used for calculating u_t)
      bar_r_t  [T-1,11] Expected return bar_r_t = Phi * r_{t-1} (estimated based on the sequence specified by use_excess)
      Sigma_r  [11,11]  Residual covariance estimation
      u_t      [T-1,11] Net USD trading volume (calculate according to MDP PDF)
      C_t      [T-1,1]  net cash flow
      B_t      [T-1,1]  Benchmark
      cash_t   [T,1]    Cash Processing (USD)
    """
    df = df_raw.copy()
    # check columns
    need = {'cusip','value','shares','filing_date'}
    if not need.issubset(df.columns):
        raise ValueError(f"df_raw need columns: {need}")
    df['value']  = pd.to_numeric(df['value'],  errors='coerce').fillna(0.0)          # convert to number
    df['shares'] = pd.to_numeric(df['shares'], errors='coerce').fillna(0.0)          # convert to number
    df['filing_date'] = pd.to_datetime(df['filing_date'])                            # convert to datetime

    # Sector Map to 11 Industries
    m = dict(sector_map[['cusip','sector']].values) if isinstance(sector_map,pd.DataFrame) else dict(sector_map)
    df['sector'] = df['cusip'].map(m).map(_to_gics_name)                              # Get standardized industry name based on mapping
    df = df.dropna(subset=['sector']).copy()                                          # Discard rows that cannot be mapped to an industry(can be adjusted)

    periods = sorted(df['filing_date'].unique().tolist())                             # Take all disclosure periods and sort them
    if len(periods) < 2:                                                              # Need at least two periods
        raise ValueError("At least two issues of 13F are required to construct the trajectory.")

    # x_t
    val_sector = (df.groupby(['filing_date','sector'])['value']                       # Aggregate value by (period, industry) (unit: thousands of US dollars)
                    .sum().unstack('sector')                                          # Transferred industries
                    .reindex(index=periods, columns=GICS_ORDER, fill_value=0.0)       # Align rows and columns to a fixed order and fill with zeros
                    .sort_index())                                                    # Sort by date
    X_dollar = (val_sector * float(value_unit)).astype(float).replace([np.inf,-np.inf], np.nan).fillna(0.0)  # Multiply units to convert to dollars and clean up exceptions
    X = torch.tensor(X_dollar.values, dtype=torch.float32, device=device)             # [T,11] Industry USD holdings tensor
    AUM = X_dollar.sum(axis=1).astype(float)                                          # [T] Total assets per period (USD)

    # ------- Separate "real raw returns for transfers" and "series for estimates/expectations" -------
    sector_px, spy_px = _fetch_sector_spy_prices(periods)                             # Download sector ETF and SPY prices

    # (1) r_t_real_raw：True "raw" industry returns (not excess), driving transfers and resolving u_t
    R_real_np = _sector_simple_returns_between_filings(                               # Calculate the raw industry returns for adjacent disclosure periods
        periods, sector_px[GICS_ORDER], spy_px, excess=False                          # excess=False means no SPY reduction
    )
    R_real = torch.tensor(R_real_np, dtype=torch.float32, device=device)              # Convert to tensor [T-1,11] as realized r_t

    # (2) R_for_model：Series to use for estimating AR(1) (optional: excess or original, controlled by use_excess)
    R_for_model_np = _sector_simple_returns_between_filings(                           # The second set of return series is only used for modeling
        periods, sector_px[GICS_ORDER], spy_px, excess=use_excess                     # True=Excess, False=Original
    )
    phi, Sigma = _estimate_ar1(R_for_model_np)                                        # Use this sequence to estimate the AR(1) coefficient and residual covariance
    phi_t = torch.tensor(phi, dtype=torch.float32, device=device)                     # Coefficients converted to tensors (11,)

    # bar_r_t = Phi * r_{t-1}
    bar_R = torch.zeros_like(R_real)
    bar_R[0] = 0.0                                                                    # The first has no previous value, so set to 0
    bar_R[1:] = torch.tensor(R_for_model_np[:-1], dtype=torch.float32, device=device) * phi_t  # Sector by sector Φ r_{t-1}

    Sigma_t = torch.tensor(Sigma, dtype=torch.float32, device=device)                 # Covariance Matrix Tensorization [11,11]

    # ------- Use "real original income" to reverse solve u_t -------
    den = (1.0 + R_real).clamp_min(1e-6)
    U = (X[1:] / den) - X[:-1]
    C = U.sum(dim=1, keepdim=True)
    B = torch.zeros((len(periods)-1, 1), dtype=torch.float32, device=device)          # Benchmark[T-1,1]

    # Cash
    cash_usd = [torch.tensor([0.0], dtype=torch.float32, device=device)]              # Initial cash (USD), set to 0 here
    cost_rate = 1e-3                                                                   # Transaction fee rate (e.g. 0.1% = 10bps)
    for t in range(len(periods)-1):                                                    # Update cash balance period by period
        traded_dollars = torch.linalg.vector_norm(U[t], ord=1)                         # Approximate transaction amount: sum of L1 norm
        trade_cost_usd = cost_rate * traded_dollars                                    # Transaction costs (USD)
        cash_next_usd = torch.clamp(cash_usd[-1] + C[t] - trade_cost_usd, min=0.0)     # Cash = Cash + Net Inflow - Cost, and is not negative
        cash_usd.append(cash_next_usd)
    CASH_USD = torch.stack(cash_usd, dim=0)                                            # [T,1]

    # Clean up(reset NaN/Inf to zero to ensure stability of subsequent training)
    X        = torch.nan_to_num(X,        nan=0.0, posinf=0.0, neginf=0.0)
    R_real   = torch.nan_to_num(R_real,   nan=0.0, posinf=0.0, neginf=0.0)
    bar_R    = torch.nan_to_num(bar_R,    nan=0.0, posinf=0.0, neginf=0.0)
    U        = torch.nan_to_num(U,        nan=0.0, posinf=0.0, neginf=0.0)
    C        = torch.nan_to_num(C,        nan=0.0, posinf=0.0, neginf=0.0)
    CASH_USD = torch.nan_to_num(CASH_USD, nan=0.0, posinf=0.0, neginf=0.0)

    return {
        "x_t": X,                                                                      # [T,11]
        "r_t": R_real,                                                                 # [T-1,11]
        "bar_r_t": bar_R,                                                              #（Φ r_{t-1})[T-1,11]
        "Sigma_r": Sigma_t,                                                            # [11,11]
        "u_t": U, "B_t": B, "C_t": C, "cash_t": CASH_USD,                               # action/benchmark/net_cash_flow/cash state(start from 0)
        "periods": periods, "sectors": GICS_ORDER,                                     # time point and industry order
        "ar1": {"phi": torch.tensor(phi)}                                              # AR(1) coefficient
    }
```

### Did they care about tracking the S&P 500? Did they care about minimizing trading costs? Did they aggressively chase growth? This code finds those answers by optimizing a set of parameters ($\rho, \eta, \lambda, w$) so that the model's calculated rewards match the actual ranking of the fund's best performance periods.

##### $\rho$ (Rho): Benchmark AdherenceRange: 0 to 1 (enforced by sigmoid).Meaning: How much does the manager "hug" the benchmark (S&P 500)?If $\rho \approx 1$: They are a passive index fund.If $\rho \approx 0$: They are an active stock picker.

##### $\eta$ (Eta): Growth TargetRange: $> 0$ (enforced by softplus).Meaning:  The aggressive growth target relative to current assets.

##### $\lambda$ (Lambda): Cash Flow SensitivityMeaning: How much does the manager dislike huge cash injections/withdrawals? A high $\lambda$ means they prefer a stable asset base.

##### $w$ (Omega): Transaction Cost AversionMeaning: How much does the manager hate paying fees?High $w$ = "Buy and Hold" strategy (rarely trades).Low $w$ = High Frequency Trader (trades constantly).

```python
def build_ranked_subtrajs(traj, min_len=2, max_windows=100): # Increased default to 100
    """
    Slice sub-trajectories and rank by Pure Portfolio Return (stripping out cash flows).
    """
    X, R = traj["x_t"], traj["r_t"]

    # --- Step 1: Calculate True Portfolio Returns (Vectorized) ---
    # We need the return of the portfolio for every single period t.
    # Formula: R_p[t] = sum(Weights[t] * SectorReturns[t])

    # 1. Calculate Weights at start of each period
    # Sum of assets per period (avoid div by zero)
    total_assets = X.sum(dim=1, keepdim=True).clamp_min(1e-9)
    weights = X / total_assets

    # 2. Calculate Portfolio Return for each period (t -> t+1)
    # R is [T-1, 11], Weights uses X[:-1] which is [T-1, 11]
    # Dot product: (Batch dot)
    port_returns = (weights[:-1] * R).sum(dim=1) # Shape: [T-1]

    # --- Step 2: Generate ALL Candidates ---
    cands = []
    T = X.shape[0]
    # Generate every possible window
    for t0 in range(0, T - min_len):
        for t1 in range(t0 + min_len - 1, T - 1):
            cands.append((t0, t1))

    # --- Step 3: Calculate Performance for ALL Candidates ---
    # We use Cumulative Product of Returns: (1+r1)*(1+r2)... - 1
    # This is much more accurate than AUM change.
    perfs = []
    valid_cands = []

    for (t0, t1) in cands:
        # Extract returns for this window
        # Window (t0, t1) covers returns from index t0 up to t1 (inclusive)
        window_rets = port_returns[t0 : t1+1]

        # Cumulative Return: Product(1 + r) - 1
        cum_ret = torch.prod(1.0 + window_rets) - 1.0

        perfs.append(cum_ret.item())
        valid_cands.append((t0, t1))

    perfs = torch.tensor(perfs, dtype=torch.float64) # Use float64 for sorting precision

    # --- Step 4: Sort & Sample ---
    # Sort ALL windows from Best to Worst
    sorted_indices = torch.argsort(perfs, descending=True)

    # Now that they are sorted, we pick a representative sample
    # We want the Best, the Worst, and even steps in between (The Staircase)
    if len(valid_cands) > max_windows:
        # Pick indices evenly spaced along the SORTED list
        # e.g., Index 0 (Best), Index 10, Index 20... Index N (Worst)
        keep_idxs = torch.linspace(0, len(valid_cands)-1, steps=max_windows).long()

        # Select the candidates
        final_indices = sorted_indices[keep_idxs]
    else:
        final_indices = sorted_indices

    # Package results
    final_cands = [valid_cands[i] for i in final_indices]
    final_perfs = perfs[final_indices]

    print(f"Generated {len(final_cands)} windows. Top Return: {final_perfs[0]:.2%}, Worst Return: {final_perfs[-1]:.2%}")

    return final_cands, final_perfs
```

##### Converts "Total Reward" into "Average Reward per Step," making a 3-month window comparable to a 16-month window.

```python
import torch
import pandas as pd

# ---------- rank sub-trajectories ----------
def build_ranked_subtrajs_old(traj, min_len=2, max_windows=30):
    """Slice sub-trajectories, rank by % AUM return."""
    X, R, U = traj["x_t"], traj["r_t"], traj["u_t"]
    aum = (X.sum(dim=1)).cpu()   # AUM per period

    cands = []
    T = X.shape[0]
    for t0 in range(0, T-1-min_len+1):
        for t1 in range(t0+min_len-1, T-2+1):
            cands.append((t0, t1))
    if len(cands) > max_windows:
        idxs = torch.linspace(0, len(cands)-1, steps=max_windows).round().long().tolist()
        cands = [cands[i] for i in idxs]

    perfs = []
    for (t0, t1) in cands:
        perf = (aum[t1+1]/(aum[t0]+1e-12) - 1.0)   # percentage return
        perfs.append(torch.tensor(float(perf)))
    perfs = torch.stack(perfs)

    order = torch.argsort(perfs, descending=True)
    return [cands[i] for i in order], perfs[order]

# ---------- reward model with normalization ----------
class ParametricReward(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.a_rho    = torch.nn.Parameter(torch.tensor(0.0))  # sigmoid -> (0,1)
        self.b_eta    = torch.nn.Parameter(torch.tensor(0.5))  # softplus -> ≥0
        self.b_lambda = torch.nn.Parameter(torch.tensor(-1.0))
        self.b_w      = torch.nn.Parameter(torch.tensor(-1.0))
        self.b_alpha  = torch.nn.Parameter(torch.tensor(0.0))

    def forward(self, traj, windows):
        X, R, U, C, B = traj["x_t"], traj["r_t"], traj["u_t"], traj["C_t"], traj["B_t"]
        # Detect the data type (Float32 or Float64) from input X
        current_dtype = X.dtype
        current_device = X.device

        # Ensure 'one' matches the data type
        one = torch.ones(X.shape[1], dtype=current_dtype, device=current_device)

        # Force parameters to match data type
        rho    = torch.sigmoid(self.a_rho).to(dtype=current_dtype)
        eta    = torch.nn.functional.softplus(self.b_eta).to(dtype=current_dtype)
        lam    = torch.nn.functional.softplus(self.b_lambda).to(dtype=current_dtype)
        w      = torch.nn.functional.softplus(self.b_w).to(dtype=current_dtype)
        alpha_p = torch.nn.functional.softplus(self.b_alpha).to(dtype=current_dtype)

        rets = []
        for (t0, t1) in windows:
            # Ensure accumulator matches data type
            r_sum = torch.tensor(0.0, dtype=current_dtype, device=current_device)

            # Calculate length of this window
            window_len = t1 - t0 + 1

            for t in range(t0, t1+1):
                AUM = X[t].sum().clamp(min=1e-6)   # scale by portfolio size

                # Normalized P and V
                Pt = (rho * B[t] + (1-rho) * eta * torch.dot(one, X[t])) / AUM
                Vt = torch.dot(1.0 + R[t], X[t] + U[t]) / AUM

                diff = (Pt - Vt) * 100.0 # Scale to Percentage points
                mismatch_cost = diff**2
                #mismatch = 100*(Pt - Vt)**2
                net_flow_cost = ((C[t].squeeze() / AUM) * 100.0)**2
                #net_flow = (((torch.sum(U[t]) - C[t].squeeze())/AUM) * 100.0)**2
                #trade_cost = torch.dot(U[t], U[t]) / (AUM**2)
                trade_cost = torch.dot(U[t], U[t]) / (AUM**2) * 10000.0 # Basis points squared
                port_ret = torch.dot(R[t], X[t]) / AUM
                alpha_bonus = alpha_p * (port_ret * 100.0)

                Rt = alpha_bonus - mismatch_cost - lam*net_flow_cost - w*trade_cost
                #Rt = - mismatch_cost - lam*net_flow_cost - w*trade_cost
                r_sum = r_sum + Rt

            avg_reward = r_sum / window_len
            rets.append(avg_reward)
        return torch.stack(rets), {
            "rho": rho.item(), "eta": eta.item(),
            "lambda": lam.item(), "w": w.item()
        }

# ---------- pairwise T-REX loss ----------
def trex_pairwise_loss(window_rewards):
    N = window_rewards.shape[0]
    loss = torch.tensor(0.0, dtype=window_rewards.dtype, device=window_rewards.device)
    cnt = 0
    for i in range(N):
        for j in range(i+1, N):
            diff = window_rewards[i] - window_rewards[j]
            loss = loss + torch.nn.functional.softplus(-diff)  # log(1+exp(-diff))
            cnt += 1
    return loss/cnt if cnt>0 else loss

# ---------- training ----------
windows, perfs_sorted = build_ranked_subtrajs(traj, min_len=2, max_windows=100)
# Cast the entire model to Double Precision to match the data
model = ParametricReward().to(traj["x_t"].device).double()
opt = torch.optim.Adam(model.parameters(), lr=0.01)

history = []
for epoch in range(1000):
    opt.zero_grad()
    rewards, params = model(traj, windows)
    loss = trex_pairwise_loss(rewards)
    loss.backward()
    opt.step()
    if (epoch+1) % 50 == 0:
        history.append((epoch+1, loss.item(), params))

# Results
df_hist = pd.DataFrame([{
    "epoch": e, "loss": l,
    "rho": p["rho"], "eta": p["eta"],
    "lambda": p["lambda"], "w": p["w"]
} for e,l,p in history])
print(df_hist)

with torch.no_grad():
    final_rewards, _ = model(traj, windows)
summary = pd.DataFrame([{
    "rank": k+1, "window": f"[{t0}->{t1+1}]",
    "perf": float(perfs_sorted[k]),
    "learned_reward": float(final_rewards[k])
} for k,(t0,t1) in enumerate(windows[:8])])
print(summary)
```

##### RL Environment

```python
class LearnedStrategyEnv(gym.Env):
    def __init__(self, traj, params):
        super(LearnedStrategyEnv, self).__init__()

        # 1. Unpack Data
        # We need Double Precision for financial math
        self.X_t = traj["x_t"].cpu().numpy().astype(np.float64) # Holdings
        self.R_t = traj["r_t"].cpu().numpy().astype(np.float64) # Market Returns
        self.B_t = traj["B_t"].cpu().numpy().astype(np.float64) # Benchmark
        self.Bar_R = traj["bar_r_t"].cpu().numpy().astype(np.float64) # Expected Returns

        self.T, self.N = self.R_t.shape

        # 2. Unpack Learned Parameters (The "Ghost")
        self.rho = params['rho']
        self.eta = params['eta']
        self.lam = params['lambda']
        self.w   = params['w']

        # 3. Define Action Space (Portfolio Weights)
        # 11 Continuous values. We will Softmax them to get % weights.
        self.action_space = spaces.Box(low=-10, high=10, shape=(self.N,), dtype=np.float64)

        # 4. Define Observation Space
        # Agent sees: [Expected_Returns (11) + Current_Weights (11)]
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(self.N * 2,), dtype=np.float64)

        # State Tracking
        self.t = 0
        self.current_portfolio_val = 1.0 # Start with $1.00
        self.current_weights = np.ones(self.N) / self.N # Start Equal Weight

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.t = 0
        self.current_portfolio_val = 1.0
        self.current_weights = np.ones(self.N) / self.N
        return self._get_obs(), {}

    def _get_obs(self):
        # Feature 1: What does the AR1 model predict for tomorrow?
        expected_ret = self.Bar_R[self.t]

        # Feature 2: How are we currently invested?
        current_alloc = self.current_weights

        return np.concatenate([expected_ret, current_alloc])

    def step(self, action):
        # 1. Action -> New Weights (Softmax)
        # e.g., Action [10, 0, 0...] -> Weights [0.99, 0.001, ...]
        exp_action = np.exp(action)
        new_weights = exp_action / np.sum(exp_action)

        # 2. Calculate "Trades" (U_t) logic
        # Trade = New_Weights - Old_Weights (Simplified for RL speed)
        # In reality, U_t depends on price moves, but for RL policy learning,
        # tracking weight changes is sufficient proxy for turnover cost.
        trade_vector = new_weights - self.current_weights
        trade_magnitude = np.sum(np.abs(trade_vector)) # L1 Norm of turnover

        # 3. Simulate Market Move
        # Portfolio Return = sum(Weights * Sector_Returns)
        realized_ret_vector = self.R_t[self.t]
        port_ret = np.dot(new_weights, realized_ret_vector)

        # Update Portfolio Value
        prev_val = self.current_portfolio_val
        self.current_portfolio_val *= (1.0 + port_ret)
        current_val = self.current_portfolio_val

        # 4. CALCULATE REWARD (Using the IRL Formula)
        # We calculate the terms exactly as the IRL model defined them.

        # Term A: Target vs Actual (The Mismatch)
        # Target = rho * Benchmark + (1-rho) * eta * Current_Val
        # Note: B_t in traj is normalized? If not, use port_ret as proxy for B.
        # Let's assume Benchmark Return is the Mean of Sector Returns (Proxy)
        bench_ret = np.mean(realized_ret_vector)

        # Growth Target Return = eta (Aggressive)
        # Benchmark Target Return = bench_ret

        # We compare RETURNS here (easier for RL stability than raw dollar values)
        target_ret = (self.rho * bench_ret) + ((1 - self.rho) * self.eta * 0.01)
        # Note: eta is usually a multiplier on AUM, here we treat it as a target return scaler

        # Simplified Mismatch for RL:
        # Reward = - (Target_Return - Actual_Return)^2
        # If Actual > Target, we clamp it (Symmetric Mimicry Logic)
        diff = (target_ret - port_ret) * 100.0
        mismatch_cost = diff**2

        # Term B: Trading Cost Penalty
        # w * (Turnover)^2
        cost_penalty = self.w * (trade_magnitude**2) * 100.0

        # Total Reward
        reward = - mismatch_cost - cost_penalty

        # 5. Advance Time
        self.t += 1
        terminated = (self.t >= self.T - 1)

        self.current_weights = new_weights # Update state

        info = {
            "portfolio_value": current_val,
            "return": port_ret,
            "turnover": trade_magnitude
        }

        return self._get_obs(), reward, terminated, False, info
```

##### Train and Backtest

```python
# 1. Package the Learned Params
learned_params = {
    'rho': final_rho,
    'eta': final_eta,
    'lambda': final_lambda,
    'w': final_w
}

# 2. Create Environments
train_env = DummyVecEnv([lambda: LearnedStrategyEnv(traj, learned_params)])

# 3. Initialize PPO Agent
# MlpPolicy is standard for vector inputs
print("Training RL Agent on Learned Rewards...")
model_ppo = PPO("MlpPolicy", train_env, verbose=1, learning_rate=0.0005, gamma=0.95)

# 4. Train
# 20k steps is usually enough for this dataset size
model_ppo.learn(total_timesteps=20000)

print("Training Complete.")

# 5. Backtest Comparison
# We will compare the RL Agent vs. Equal Weight Portfolio
print("\n--- STARTING BACKTEST ---")
test_env = LearnedStrategyEnv(traj, learned_params)
obs, _ = test_env.reset()
done = False

rl_wealth = [1.0]
eq_wealth = [1.0]

while not done:
    # Get RL Action
    action, _ = model_ppo.predict(obs)
    obs, reward, done, _, info = test_env.step(action)
    rl_wealth.append(info['portfolio_value'])

    # Calculate Equal Weight Benchmark (1/11th in each)
    # Re-calculate market return for this step
    t_now = test_env.t - 1
    mkt_ret = np.mean(test_env.R_t[t_now])
    eq_wealth.append(eq_wealth[-1] * (1.0 + mkt_ret))

# 6. Plot Results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(rl_wealth, label='Rebellion RL Agent', color='blue', linewidth=2)
plt.plot(eq_wealth, label='Equal Weight Benchmark', color='gray', linestyle='--')
plt.title("RL Agent Performance (Based on IRL Strategy)", fontsize=16)
plt.ylabel("Portfolio Value ($)")
plt.xlabel("Time (Periods)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Final RL Wealth: ${rl_wealth[-1]:.2f}")
print(f"Final EQ Wealth: ${eq_wealth[-1]:.2f}")
```
